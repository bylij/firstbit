### 问题2
- 已知 $\frac{dV}{dt}=30$  
$V=\frac{1}{3} \cdot \pi \cdot (\frac{h}{2})^2 \cdot h=\frac{1}{12}\pi h^3 \implies \frac{dV}{dt}=\frac{\pi h^2}{4} \cdot \frac{dh}{dt}\\
\implies \frac{dh}{dt}=\frac{dV}{dt} \cdot \frac{4}{\pi h^2}=30 \cdot \frac{4}{\pi \cdot 10^2} \approx 0.38$ 英尺/分

### 问题4
- 设圆柱半径和高分别为 $r, h$  
已知限制条件 $V=\pi r^2h+\frac{2}{3}\pi r^3 \implies (V)^\prime=(\pi r^2h)^\prime+(\frac{2}{3}\pi r^3)^\prime \implies 0=\pi(2rh+r^2h^\prime)+2\pi r^2 \implies h^\prime=-\frac{2r+2h}{r}$ $\textcircled{1}$  
表面积 $S=\pi r^2+2\pi rh+2\pi r^2=3\pi r^2+2\pi rh \implies S^\prime=(3\pi r^2)^\prime+(2\pi rh)^\prime=6\pi r+2\pi(h+rh^\prime)$  
带入 $\textcircled{1}$ 得 $S^\prime=6\pi r+2\pi(h-2r-2h)=2\pi(r-h)$ 函数在 $r=h$ 时取得极点  
限制条件 $V=\pi r^2h+\frac{2}{3}\pi r^3 \implies h=\frac{V-\frac{2}{3}\pi r^3}{\pi r^2}$ 带入到 $S$ 得 $S=\frac{5}{3}\pi r^2+\frac{2V}{r}$ 其中 $0 \lt r \lt \infty$  
验证 $S(r=0^+)=\infty, S(r=\infty)=\infty$，因此函数 $S$ 在极点处取得最小值，即 $r=h$ 时满足条件  
可得 $r=h=\sqrt[3]{\frac{3v}{5\pi}}$ 时所用金属材料最少

### 问题5
- $f(x)=x^3-3x+7 \implies f^\prime(x)=3x^2-3$  
根据牛顿法 $x_1=2 \implies x_2=x_1-\frac{f(x_1)}{f^\prime(x_1)}=2-\frac{2^3-3 \cdot 2+7}{3 \cdot 2^2-3}=1$  
可得 $f^\prime(x_2)=3 \cdot 1^2-3=0$ 这导致无法进行下一轮迭代  
而此时 $f(x_2=1)=1^3-3 \cdot 1+7=5$ 仍与 $f(x)=0$ 相差较大，因此 $x_1=2$ 这个初始值使得牛顿法不能很好的进行

### 问题6
- $f(x)=\sqrt{1+x}-(1+\frac{1}{2}x) \implies f^\prime(x)=\frac{1}{2\sqrt{1+x}}-\frac{1}{2}=\frac{1}{2}(\frac{1}{\sqrt{1+x}}-1)=\frac{1-\sqrt{1+x}}{2\sqrt{1+x}}$  
$x \gt 0 \implies \sqrt{1+x} \gt 1 \implies f^\prime(x) \lt 0$ 得函数是递减函数  
因此 $x \gt 0 \implies f(x) \lt f(0) \implies \sqrt{1+x}-(1+\frac{1}{2}x) \lt 0 \implies \sqrt{1+x} \lt 1+\frac{1}{2}x$
