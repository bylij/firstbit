- b) 设 $u=x^2 \implies x=\pm \sqrt{u}$  
取 $x=\sqrt{u} \implies \int_0^u{f(t)dt}=\sqrt{u}\sin{(\pi\sqrt{u})}$ $\textcircled{1}$  
取 $x=-\sqrt{u} \implies \int_0^u{f(t)dt}=-\sqrt{u}\sin{(\pi \cdot -\sqrt{u})}=\sqrt{u}\sin{(\pi\sqrt{u})}$ $\textcircled{2}$  
综合 $\textcircled{1}$ 和 $\textcircled{2}$ 可得 $\int_0^u{f(t)dt}=\sqrt{u}\sin{(\pi\sqrt{u})} \implies f(u)=\frac{d}{du}\int_0^u{f(t)dt}=\frac{d}{du}\sqrt{u}\sin{(\pi\sqrt{u})}=\frac{1}{2\sqrt{u}} \cdot \sin{(\pi\sqrt{u})}+\sqrt{u} \cdot \cos{(\pi\sqrt{u})} \cdot \frac{\pi}{2\sqrt{u}}=\frac{\sin{(\pi\sqrt{u})}}{2\sqrt{u}}+\frac{\pi}{2}\cos{(\pi\sqrt{u})}$  
可得 $f(4)=\frac{\sin{(2\pi)}}{4}+\frac{\pi}{2}\cos{(2\pi)}=0+\frac{\pi}{2} \cdot 1=\frac{\pi}{2}$

### 问题6
- a) 设盐的质量随时间变化的函数 $m=m(t)$，液体的质量 $M$，已知水的密度为 $1 kg/L$  
可得在 $t$ 时刻液体的密度 $\rho(t)=\frac{1000 \cdot 1+m(t)}{1000}=\frac{1000+m(t)}{1000}$，在 $t$ 时刻液体的质量分数 $w(t)=\frac{m(t)}{1000+m(t)}$  
根据液体流出速度为 10L/min 可得 $10 \cdot dt \cdot \rho(t) \cdot w(t)=-dm \implies 10 \cdot dt \cdot \frac{1000+m(t)}{1000} \cdot \frac{m(t)}{1000+m(t)}=-dm \implies \frac{dm}{dt}=-\frac{1}{100}m(t)$
- b) 根据微分方程 $\frac{dm}{dt}=-\frac{1}{100}m \implies \frac{dm}{m}=-\frac{1}{100}dt$ 两侧同时求不定积分 $\int{\frac{dm}{m}}=\int{-\frac{1}{100}dt} \implies \ln{m}=-\frac{1}{100}t+C$ 其中 $C$ 为常量  
由 $\ln{m}=-\frac{1}{100}t+C \implies m=e^{-\frac{1}{100}t+C}=Ae^{-\frac{1}{100}t}$ 其中 $A$ 为常量  
由 $m(t=0)=15 \implies Ae^0=15 \implies A=15$ 即液体中盐分质量随时间变化的方程为 $m(t)=15e^{-\frac{1}{100}t}$
- c) 由 $m(t)=\frac{15}{2} \implies 15e^{-\frac{1}{100}t}=\frac{15}{2} \implies e^{-\frac{1}{100}t}=\frac{1}{2} \implies -\frac{1}{100}t=\ln{\frac{1}{2}}$ 解得 $t=100\ln{2} \approx 69.3$ 分钟
