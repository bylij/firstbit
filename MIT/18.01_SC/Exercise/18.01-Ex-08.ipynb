### 问题
- 1、函数是一个偶函数，函数曲线相对 $y$ 轴对称  
2、函数在 $x=0$ 点无定义，但是根据 $\lim\limits_{x \to 0}\frac{\sin{x}}{x}=1$ 这个条件可以描绘 $x=0$ 点附近的曲线  
3、除了 $x=0$ 点外，函数曲线与 $x$ 轴的交点与 $\sin{x}$ 相同  
4、函数具有类似于 $\sin{x}$ 的震荡性，可以看成是一个振幅不断变小的类 $\sin{x}$ 曲线，$|\frac{1}{x}|$ 就是这个变动的振幅
