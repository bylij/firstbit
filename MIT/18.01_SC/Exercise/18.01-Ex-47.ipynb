### 问题
- a) $\int_0^2{x^2dx}=\frac{x^3}{3}|_0^2=\frac{2^3}{3}-0=\frac{8}{3}$  
几何示意图如下，积分即为阴影部分的面积  
  
采用黎曼和进行近似计算，将 $[0, 2]$ 区间六等分，$\Delta{x}=\frac{2-0}{6}=\frac{1}{3}$，其中 $f(x)=x^2$  
阴影部分面积 $S \approx \Sigma_{i=0}^5{f(\frac{i\Delta{x}+(i+1)\Delta{x}}{2})\Delta{x}}=\Sigma_{i=0}^5{(\frac{\frac{i}{3}+\frac{i+1}{3}}{2})^2\frac{1}{3}}=\frac{1}{108}\Sigma_{i=0}^5{(2i+1)^2}=\frac{286}{108} \approx 2.64$ 这与 $\frac{8}{3} \approx 2.66$ 是很接近的
- b) $\int_1^e{\frac{1}{x}dx}=\ln{x}|_1^e=\ln{e}-\ln{1}=1-0=1$，同样可以采用黎曼和近似计算几何面积法来求解，这里省略
- c) $\int_{\frac{\pi}{4}}^0{\sin{x}dx}=(-\cos{x})|_{-\frac{\pi}{4}}^0=-\cos{0}-(-\cos{(-\frac{\pi}{4})})=-1-(-\frac{\sqrt{2}}{2})=\frac{\sqrt{2}}{2}-1$，这里省略黎曼法近似计算几何面积的部分
