- a) 定积分 $\int_0^\pi{(\sin{x}+\cos{x})dx}$ 为正弦和余弦曲线在 $[0, \pi]$ 区间内与 $x$ 轴组成的图形面积之和，如下图：  
  
其中余弦曲线的图形分属于 $x$ 轴上下，面积抵消，因此仅剩下正弦曲线与 $x$ 轴形成的面积，即 $\int_0^\pi{(\sin{x}+\cos{x})dx}=\int_0^\pi{\sin{x}dx}$  
采用黎曼和进行近似计算，将 $[0, \pi]$ 区间进行六等分，$\Delta{x}=\frac{\pi}{6} \implies \int_0^\pi{\sin{x}dx} \approx \Sigma_{i=0}^5{\sin{\frac{i\Delta{x}+(i+1)\Delta{x}}{2}}\Delta{x}}=\frac{\pi}{6}(\sin{15^\circ}+\sin{45^\circ}+\sin{75^\circ}+\sin{105^\circ}+\sin{135^\circ}+\sin{165^\circ}) \approx 2.02$
- b) $\int{\cos{x}dx}=\sin{x}+C, \int{\sin{x}dx}=-\cos{x}+C$ 其中 $C$ 为常量
- c) $\int_0^\pi{(\sin{x}+\cos{x})dx}=\int_0^\pi{\sin{x}dx}+\int_0^\pi{\cos{x}dx}\\
=(-\cos{x}|_0^\pi)+\sin{x}|_0^\pi\\
=(-\cos{\pi}-(-\cos{0}))+(\sin{\pi}-\sin{0})\\
=(-(-1)-(-1))+(0-0)\\
=2$
