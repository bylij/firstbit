### 问题1C-1
- a) $S=\pi r^2$  
$\frac{dS}{dr}=\lim\limits_{\Delta{r} \to 0}\frac{2\pi (r+\Delta{r})^2-2\pi r}{\Delta{r}}\\
=2\pi \cdot \lim\limits_{\Delta{r} \to 0}\frac{r^2+2r \cdot \Delta{r}+\Delta{r}^2-r^2}{\Delta{r}}\\
=2\pi \cdot \lim\limits_{\Delta{r} \to 0}(2r+\Delta{r})\\
=2\pi r$

### 问题1C-4
- a) $f^\prime(x)=-\frac{2}{(2x+1)^2}$，可得 $f^\prime(1)=-\frac{2}{9}$ 为切线斜率  
而 $f(1)=\frac{1}{2 \cdot 1+1}=\frac{1}{3}$，即切线过 $(1, \frac{1}{3})$ 点  
可得切线的点斜公式 $y-\frac{1}{3}=-\frac{2}{9}(x-1)$
- b) $f^\prime(x)=4x+5$，可得 $f^\prime(a)=4a+5$ 为切线斜率  
而 $f(a)=2a^2+5a+4$，即切线过 $(a, 2a^2+5a+4)$ 点  
可得切线公式 $y-(2a^2+5a+4)=(4a+5)(x-a)$

### 问题1C-5
