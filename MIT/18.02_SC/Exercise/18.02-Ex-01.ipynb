### 问题3
- 如下图以向量 $\overrightarrow{OA}$ 和 $\overrightarrow{OB}$ 构成的平行四边形  
  
设 P点为 AB 的中点，Q点为 OC 的中点，只要证明 P点与 Q点为同一点即可证明命题成立  
已知 $\overrightarrow{OC}=\overrightarrow{OA}+\overrightarrow{OB}$，$\overrightarrow{AB}=\overrightarrow{OB}-\overrightarrow{OA}$ 而 $\overrightarrow{AP}=\frac{1}{2}\overrightarrow{AB}$ 且 $\overrightarrow{OQ}=\frac{1}{2}\overrightarrow{OC}$  
可得 $\overrightarrow{OP}=\overrightarrow{OA}+\overrightarrow{AP}=\overrightarrow{OA}+\frac{1}{2}\overrightarrow{AB}=\overrightarrow{OA}+\frac{1}{2}(\overrightarrow{OB}-\overrightarrow{OA})=\frac{1}{2}(\overrightarrow{OB}+\overrightarrow{OA})=\frac{1}{2}\overrightarrow{OC}=\overrightarrow{OQ}$  
由 $\overrightarrow{OP}=\overrightarrow{OQ}$ 可得 P点与Q点重合，即为同一个点，命题成立

### 问题4
- 如下图 ABCD 为空间四边形，其中 P，Q，M，N 分别为四个边的中点  
  
$\overrightarrow{PQ}=\overrightarrow{PB}+\overrightarrow{BQ}=\frac{1}{2}(\overrightarrow{AB}+\overrightarrow{BC})=\frac{1}{2}\overrightarrow{AC}$  
$\overrightarrow{NM}=\overrightarrow{ND}+\overrightarrow{DM}=\frac{1}{2}(\overrightarrow{AD}+\overrightarrow{DC})=\frac{1}{2}\overrightarrow{AC}$  
因此 $\overrightarrow{PQ}=\overrightarrow{NM} \implies PQ//NM$，同理得 $NP//MQ$，命题成立
