### 问题1
- $\vec{A}=<1, 8>, \vec{B}=<1, 2>$  
可得 $\cos{\theta}=\frac{\vec{A} \cdot \vec{B}}{|\vec{A}||\vec{A}|}=\frac{1 \cdot 1+8 \cdot 2}{\sqrt{1^2+8^2} \cdot \sqrt{1^2+2^2}}=\frac{17}{5\sqrt{13}} \implies \theta=\cos^{-1}\frac{17}{5\sqrt{13}} \approx 19.47^\circ$

### 问题2
- 由已知得 $\overrightarrow{QP}=<a, 0, -2>, \overrightarrow{QR}=<a, -2, 2>$  
$\angle{PQR}=90^\circ \implies \overrightarrow{QP} \cdot \overrightarrow{QR}=a^2+0+(-2) \cdot 2=0 \implies a= \pm 2$

### 问题3
- 如下图平行四边形 ABDC，其中 $AD \perp BC$  
  
$AD \perp BC \implies \overrightarrow{AD} \cdot \overrightarrow{BC}=0\\
\implies (\overrightarrow{AB}+\overrightarrow{AC}) \cdot (\overrightarrow{AB}-\overrightarrow{AC})=0\\
\implies \overrightarrow{AB} \cdot \overrightarrow{AB}-\overrightarrow{AC} \cdot \overrightarrow{AC}=0\\
\implies |\overrightarrow{AB}|^2-|\overrightarrow{AC}|^2=0\\
\implies |\overrightarrow{AB}|=|\overrightarrow{AC}|$  
可得 AB 与 AC 边的长度相等，由于 ABDC为平行四边形，因此可得其四个边均相等，因此 ABDC 为菱形
