### 问题1
- 示意图如下  
  
$\overrightarrow{OP}=\overrightarrow{OA}+\overrightarrow{AP}$  
$\overrightarrow{OA}=<12\sqrt{2}t,12\sqrt{2}t>$  
$\overrightarrow{AP}=<2\cos{\theta},2\sin{\theta}>=<2\cos{(6\pi t)},2\sin{(6\pi t)}>$  
得参数方程 $\begin{cases}x(t)=12\sqrt{2}t+2\cos{(6\pi t)}\\y(t)=12\sqrt{2}t+2\sin{(6\pi t)}\end{cases}$
