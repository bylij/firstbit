### 问题
- 问题示意图如下：  
  
$Area(R)=\iint_R{dxdy}$ $\textcircled{1}$ 可知区域 $R$ 范围：$\frac{1}{x} \le y \le \frac{2}{x} \implies 1 \le xy \le 2$ 且 $x^2 \le y \le 2x^2 \implies \frac{1}{2} \le \frac{x^2}{y} \le 1$  
设 $u=\frac{x^2}{y}, v=xy \implies J=\frac{\partial(u, v)}{\partial(x, y)}=\begin{vmatrix}u_x&&u_y\\v_x&&v_y\end{vmatrix}=\begin{vmatrix}\frac{2x}{y}&&-\frac{x^2}{y^2}\\y&&x\end{vmatrix}=\frac{3x^2}{y} \implies dudv=|J|dxdv \implies dxdy=\frac{1}{|J|}dudv=\frac{y}{3x^2}dudv=\frac{1}{3u}dudv$  
带入 $\textcircled{1}$ 可得 $Area(R)=\iint_R{dxdy}=\int_1^2{\int_{\frac{1}{2}}^1{\frac{1}{3u}du}dv}=\int_1^2{\left[\frac{1}{3}\ln{u}\right]_{\frac{1}{2}}^1dv}=\int_1^2{\frac{\ln{2}}{3}dv}=\left[\frac{\ln{2}}{3}v\right]_1^2=\frac{\ln{2}}{3}$
