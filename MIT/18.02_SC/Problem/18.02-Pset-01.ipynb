### 问题1A-8
- a) $\vec{A}$ 在 $\hat{i}$ 的分量 $a=\vec{A} \cdot \hat{i}=|\vec{A}|\cos{\alpha} \implies \cos{\alpha}=\frac{a}{|\vec{A}|}$，同理可得 $\cos{\beta}=\frac{b}{|\vec{A}|}, \cos{\gamma}=\frac{c}{|\vec{A}|}$  
而 $dir(\vec{A})=\frac{\vec{A}}{|\vec{A}|}=\frac{a\hat{i}+b\hat{j}+c\hat{k}}{|\vec{A}|}=\frac{a}{|\vec{A}|}\hat{i}+\frac{b}{|\vec{A}|}\hat{j}+\frac{c}{|\vec{A}|}\hat{k}=\cos{\alpha}\hat{i}+\cos{\beta}\hat{j}+\cos{\gamma}\hat{k}$
- b) $\cos{\alpha}=\frac{a}{|\vec{A}|}=\frac{a}{\sqrt{a^2+b^2+c^2}}, \cos{\beta}=\frac{b}{|\vec{A}|}=\frac{b}{\sqrt{a^2+b^2+c^2}}, \cos{\gamma}=\frac{c}{|\vec{A}|}=\frac{c}{\sqrt{a^2+b^2+c^2}}$  
向量 $-\hat{i}+2\hat{j}+2\hat{k}$ 的方向余弦值 $\cos{\alpha}=-\frac{1}{3}, \cos{\beta}=\frac{2}{3}, \cos{\gamma}=\frac{2}{3}$
- c) 如果 $t, u, v$ 是向量 $\vec{A}$ 的方向余弦，则 $t\hat{i}+u\hat{j}+v\hat{k}=dir(\vec{A})$ 是一个单位向量，因此 $\sqrt{t^2+u^2+v^2}=1 \implies t^2+u^2+v^2=1$  
反过来 $t^2+u^2+v^2=1$ 则 $t\hat{i}+u\hat{j}+v\hat{k}=\vec{u}$ 是一个单位向量，$t, u, v$ 是向量 $\vec{u}$ 的方向余弦

### 问题1A-11
- 如下图平行四边形 ABCD，X为 AC 中点，Y为 BD 中点  
  
要证明 AC 与 BD 互相等分，只需要证明 X点与 Y点重合即可  
$\overrightarrow{AY}=\overrightarrow{AB}+\overrightarrow{BY}=\overrightarrow{AB}+\frac{1}{2}\overrightarrow{BD}=\overrightarrow{AB}+\frac{1}{2}(\overrightarrow{BA}+\overrightarrow{BC})=\frac{1}{2}(\overrightarrow{BC}-\overrightarrow{BA})=\frac{1}{2}\overrightarrow{AC}=\overrightarrow{AX}$  
因此，点X与点Y重合，命题成立
