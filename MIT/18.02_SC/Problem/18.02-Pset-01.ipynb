### 问题1B-13
- 如下图所示 $\vec{u_1}, \vec{u_2}$ 是以原点为起点，单位圆上点为终点的两个原点向量，与 $x$轴夹角分别为 $\theta_1, \theta_2$  
  
$\vec{u_1} \cdot \vec{u_2}=|\vec{u_1}||\vec{u_2}|\cos{(\theta_1-\theta_2)}=\cos{(\theta_1-\theta_2)}$ $\textcircled{1}$  
假设 $\vec{u_1}=<a_1,a_2>, \vec{u_2}=<b_1,b_2>$  
$\vec{u_1} \cdot \hat{i}=|\vec{u_1}||\hat{i}|\cos{\theta_1}=\cos{\theta_1} \implies \cos{\theta_1}=\vec{u_1} \cdot \hat{i}=a_1$ 同理得 $\cos{\theta_2}=b_1$  
$\vec{u_1} \cdot \hat{j}=|\vec{u_1}||\hat{j}|\cos{90^{\circ}-\theta_1}=\sin{\theta_1} \implies \sin{\theta_1}=\vec{u_1} \cdot \hat{j}=a_2$ 同理得 $\sin{\theta_2}=b_2$  
由此 $\cos{\theta_1}\cos{\theta_2}+\sin{\theta_1}\sin{\theta_2}=a_1b_1+a_2b_2=\vec{u_1} \cdot \vec{u_2}$ $\textcircled{2}$  
结合 $\textcircled{1}$，$\textcircled{2}$ 可得 $\cos{(\theta_1-\theta_2)}=\cos{\theta_1}\cos{\theta_2}+\sin{\theta_1}\sin{\theta_2}$

### 问题1B-10
- 如下图 ABCD 为平行四边形，其中 AC 和 BD 长度相等  
  
$|\overrightarrow{AC}|=|\overrightarrow{BD}| \implies |\overrightarrow{AC}|^2=|\overrightarrow{BD}|^2\\
\implies \overrightarrow{AC} \cdot \overrightarrow{AC}=\overrightarrow{BD} \cdot \overrightarrow{BD}\\
\implies (\overrightarrow{AD}+\overrightarrow{AB}) \cdot (\overrightarrow{AD}+\overrightarrow{AB})=(\overrightarrow{AD}-\overrightarrow{AB}) \cdot (\overrightarrow{AD}-\overrightarrow{AB})\\
\implies \overrightarrow{AD} \cdot \overrightarrow{AD}+\overrightarrow{AB} \cdot \overrightarrow{AB}+2\overrightarrow{AD} \cdot \overrightarrow{AB}=\overrightarrow{AD} \cdot \overrightarrow{AD}+\overrightarrow{AB} \cdot \overrightarrow{AB}-2\overrightarrow{AD} \cdot \overrightarrow{AB}\\
\implies \overrightarrow{AD} \cdot \overrightarrow{AB}=0\\
\implies \overrightarrow{AD} \perp \overrightarrow{AB}$  
因此平行四边形 ABCD 为矩形

### 问题1B-11
- 如下图 ABCD 为平行四边形  
  
$\overrightarrow{AC} \perp \overrightarrow{BD} \iff \overrightarrow{AC} \cdot \overrightarrow{BD}=0\\
\iff (\overrightarrow{AB}+\overrightarrow{AD}) \cdot (\overrightarrow{AD}-\overrightarrow{AB})=0\\
\iff \overrightarrow{AD} \cdot \overrightarrow{AD} - \overrightarrow{AB} \cdot \overrightarrow{AB}=0\\
