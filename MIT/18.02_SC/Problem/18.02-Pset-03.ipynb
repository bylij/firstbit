### 问题1K-3
- 中心力运动 $\vec{a} // \vec{r} \implies \vec{a} \times \vec{r}=0 \implies \frac{d\vec{v}}{dt} \times \vec{r}=0$ $\textcircled{1}$  
$\frac{d\vec{r}}{dt} \times \vec{v}=\vec{v} \times \vec{v}=0$ 结合 $\textcircled{1}$ 可得 $\frac{d\vec{r}}{dt} \times \vec{v}+\frac{d\vec{v}}{dt} \times \vec{r}=0 \implies \frac{d}{dt}(\vec{r} \times \vec{v})=0$  
可得 $\vec{r} \times \vec{v}$ 是常向量  
由于 $\vec{r} \perp \vec{r} \times \vec{v}$ 且 $\vec{v} \perp \vec{r} \times \vec{v}$，$\vec{r} \times \vec{v}$ 是常向量则 $\vec{r} \times \vec{v}$ 方向不变，因此 $\vec{r}$ 和 $\vec{v}$ 将在保持在一个以 $\vec{r} \times \vec{v}$ 为法向量的平面内，即该运动轨迹保持在一个平面内 $\textcircled{a}$  
运动扫过的面积 $S=\frac{1}{2}|\vec{r} \times \vec{v}|$ 由于 $\vec{r} \times \vec{v}$ 是常向量，可得 $S=\frac{1}{2}|\vec{r} \times \vec{v}|$是常数 $\textcircled{b}$  
结合  $\textcircled{a}$ 和 $\textcircled{b}$ 可证命题成立
