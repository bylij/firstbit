### 问题2H-6
- a) 已知 $x+y+z=4$  
矩形和正方形总面积 $S=xy+z^2/4=xy+\frac{1}{4}[4-(x+y)]^2 \implies \begin{cases}S_x=\frac{x+3y}{2}-2\\S_y=\frac{3x+y}{2}-2\end{cases}$  
由 $S_x=S_y=0 \implies \begin{cases}\frac{x+3y}{2}-2=0\\\frac{3x+y}{2}-2=0\end{cases}$ 解得驻点 $(1, 1)$ 带入得 $S=1 \cdot 1+\frac{1}{4}(4-2)^2=2$  
函数的定义域边界为：左边界 $x=0, y \in (0, 4)$ 下边界 $y=0, x \in (0, 4)$ 右上边界 $y=4-x, x \in (0, 4)$ 是一个三角形边界  
（一）左边界验证：将 $\begin{cases}x=0\\y=4t,(t \in (0, 1))\end{cases}$ 带入原方程得 $S=4(t-1)^2$，极大值 $S(t=0^+)=4$，极小值 $S(t=1^-)=0$  
（二）下边界验证：将 $\begin{cases}y=0\\x=4s,(s \in (0, 1))\end{cases}$ 带入原方程得 $S=4(s-1)^2$，极大值 $S(s=0^+)=4$，极小值 $S(s=1^-)=0$
